# Transponieren von Daten in Power BI (Transpose)

Das **Transponieren** von Daten (englisch: *Transpose*) ist ein wichtiger Schritt in der Datenaufbereitung, bei dem Zeilen und Spalten einer Tabelle vertauscht werden. In Power BI geschieht dieser Vorgang typischerweise im **Power Query Editor**. Durch das Transponieren lassen sich Tabellen, die ursprünglich zeilenorientiert aufgebaut wurden, in eine spaltenorientierte Struktur überführen – oder umgekehrt.

---

## Begriffserklärung und Bedeutung

- **Transponieren**: Vertauschen von Zeilen und Spalten, sodass die erste Zeile zu einer neuen Spaltenüberschrift wird und jede Spalte einer Zeile in eine Zeile überführt wird (bzw. umgekehrt).
- **Relevanz**: Das Transponieren ist besonders dann hilfreich, wenn Daten in einem unüblichen Format vorliegen, das für Analysen in Power BI ungeeignet ist. Durch das Vertauschen von Zeilen und Spalten gewinnst du eine flexiblere und oftmals besser strukturierte Datengrundlage.

---

## Was ist Transponieren?

Transponieren bedeutet, eine Tabelle so umzubauen, dass die **Zeilen zu Spalten** werden und die **Spalten zu Zeilen**. Stell dir eine Matrix in der Mathematik vor: Das **Transpose** einer Matrix kehrt deren Struktur genau um. Dasselbe Prinzip gilt für Tabellen im Power Query Editor.

---

## Warum transponieren?

1. **Formatierung anpassen**: Oft liegen Daten in einem Format vor, das mehr für die **Präsentation** als für die **Analyse** gedacht ist (z. B. Überschriften in der ersten Spalte statt im Tabellenkopf). Durch Transponieren erhältst du eine analytisch besser nutzbare Struktur.
2. **Datenmodellierung vereinfachen**: Manche Analysen erfordern, dass bestimmte Felder als Spaltenüberschriften verwendet werden oder umgekehrt. Ein Transpose-Schritt kann hier die nachfolgende Modellierung deutlich vereinfachen.
3. **Flexibilität bei Berichten**: Bei stark zeilenorientierten Daten hilft es gelegentlich, diese in spaltenorientierte Daten zu verwandeln, damit Power BI sie leichter verarbeiten und visualisieren kann.

---

## Wann wird transponiert?

- **Ungeeignete Tabellenstruktur**: Wenn die aktuellen Spaltenüberschriften eigentlich als Zeilenwerte gedacht sind (z. B. Monate, Produkte, Kategorien).
- **Wiederkehrende Überschriften in Zeilen**: Etwa wenn in der Originalquelle jeder Datenblock neu überschrieben wurde, statt Spaltenüberschriften nur einmal oben anzugeben.
- **Vorbereitende Umwandlung**: Vor anderen Schritten wie **Pivot** oder **Unpivot** kann ein Transpose erforderlich sein, um die Daten in eine sinnvolle Ausgangsform zu bringen.

---

## Wo und wie transponiert man in Power BI?

Das **Transponieren** führst du im **Power Query Editor** aus. So gehst du vor:

1. **Daten importieren**: Lade die gewünschte Datei oder Datenquelle in Power BI und öffne den Power Query Editor über **„Daten transformieren“**.
2. **Tabelle auswählen**: Wähle die Abfrage (Tabelle) aus, die du transponieren möchtest.
3. **Transformieren → Transponieren**: Im Menüband im **Reiter „Transformieren“** findest du die Schaltfläche **„Transponieren“**. Klicke darauf, um die Zeilen und Spalten zu vertauschen.
4. **Kopfzeile verwenden (optional)**: Wenn deine transponierten Daten Überschriften haben, wähle nach dem Transponieren **„Erste Zeile als Kopfzeile verwenden“**. So erhöhst du die Lesbarkeit und Struktur.

---

## Vorteile des Transponierens

1. **Bessere Datenstruktur**  
   Durch das Vertauschen von Zeilen und Spalten erhältst du häufig eine Tabellenstruktur, die in Power BI wesentlich einfacher verarbeitet und analysiert werden kann.

2. **Schnelle Lösung für formatierte Quellen**  
   In vielen Fällen liegen Daten beispielsweise aus Excel in einem layoutorientierten Format vor. Ein schneller Transpose-Schritt kann sie in ein analysetaugliches Format konvertieren, ohne dass man umfangreiche manuelle Anpassungen vornehmen muss.

3. **Verbessertes Datenverständnis**  
   Du kannst die Daten gezielt so umformen, dass die Beziehung zwischen verschiedenen Feldern klarer erkennbar wird. Gerade bei Berichten, in denen „Monate“ als Spalten statt als Zeilen angelegt wurden, unterstützt ein Transpose die nachfolgende Datenmodellierung oder Visualisierung enorm.

4. **Flexible Weiterverarbeitung**  
   Einmal transponierte Daten lassen sich einfacher **pivotisieren** oder **entpivotisieren**, falls das nötig sein sollte. Oft dient Transponieren als Zwischenschritt bei komplexen Transformationen.

---

## Fazit

Das **Transponieren von Daten** ist eine vergleichsweise einfache, aber wirkungsvolle Funktion im Power Query Editor von Power BI. Mit wenigen Klicks lassen sich **Zeilen und Spalten vertauschen**, wodurch aus einer schwer analysierbaren Datentabelle eine deutlich strukturiertere Form wird. Ob du Daten aus Excel, CSV oder anderen Quellen erhältst – wenn deren Layout nicht ideal für deine Analyse ist, kann das **Transponieren** dir viel Arbeit in der nachfolgenden Modellierung und Visualisierung ersparen.
