## Theorie: Datenquellen in Power BI

Um Berichte und Dashboards in Power BI zu erstellen, benötigt man zunächst Daten. Power BI erlaubt den Zugriff auf eine Vielzahl von Datenquellen, die in unterschiedliche Kategorien eingeordnet werden können. Das Verständnis dieser Datenquellen und ihrer Einsatzmöglichkeiten ist entscheidend für die effiziente Nutzung von Power BI.

### Kategorien von Datenquellen

Power BI unterstützt mehrere Arten von Datenquellen, die wir wie folgt kategorisieren:

#### 1. Datei

Diese Kategorie umfasst verschiedene Dateitypen, die lokal oder auf Cloudspeichern gespeichert sein können. Dazu zählen:

- **Excel-Arbeitsmappen (.xlsx)**: Häufig genutzt aufgrund ihrer einfachen Handhabung und weiten Verbreitung.
- **CSV-Dateien (.csv)**: Sehr wichtig und oft verwendet, besonders im Bereich Data Science und Analytics.
- **Textdateien (.txt)**: Nützlich für einfache tabellarische Daten.
- **XML-Dateien (.xml)**: Relevant für Daten, die strukturiert hierarchisch gespeichert sind.
- **JSON-Dateien (.json)**: Beliebt aufgrund ihrer leichten Integration mit Webanwendungen.
- **PDF-Dateien (.pdf)**: Praktisch, wenn Daten aus Berichten oder Dokumenten extrahiert werden sollen.

CSV-Dateien nehmen hierbei eine zentrale Rolle ein, da sie sehr einfach und universell verwendbar sind, und häufig als Standardformat für Datenaustausch dienen.

#### 2. Datenbanken

Power BI ermöglicht direkte Verbindungen zu Datenbanken, was den Zugriff auf strukturierte Daten erleichtert. Beispiele hierfür sind:

- **SQL Server Datenbanken**: Sehr populär, da sie schnelle und komplexe Abfragen ermöglichen.
- **Access Datenbanken**: Eignen sich für kleinere, lokal gespeicherte Datenbanken.

Darüber hinaus bietet Power BI Anbindungen zu vielen weiteren Datenbanksystemen (z.B. Oracle, MySQL, PostgreSQL), auf die im fortgeschrittenen Verlauf eingegangen werden kann.

#### 3. Microsoft Fabric Semantische Modelle

Microsoft Fabric ermöglicht die Erstellung semantischer Modelle, die dann direkt in Power BI genutzt werden können. Diese Modelle helfen, komplexe Datenlandschaften zu vereinfachen und zentralisiert bereitzustellen.

#### 4. Power Platform

Power BI integriert sich nahtlos in die Microsoft Power Platform. Wichtige Beispiele hierfür sind:

- **Power BI Dataflows**: Ermöglichen die Wiederverwendung und Vereinfachung von Datenvo

## Strukturierte vs. Unstrukturierte Daten in Power BI

Bei der Arbeit mit Power BI unterscheidet man grundsätzlich zwischen **strukturierten** und **unstrukturierten Datenquellen**. Diese Unterscheidung ist wichtig, da sie großen Einfluss auf die Schritte der Datenaufbereitung, Modellierung und Visualisierung hat.

Power BI benötigt für die Analyse immer strukturierte Daten, also Daten, die bereits in einer klar definierten, tabellarischen Form vorliegen. Beispiele dafür sind Tabellen in Datenbanken oder Excel-Dateien.

Unstrukturierte Daten, wie beispielsweise CSV- oder Textdateien, besitzen dagegen keine festgelegte Struktur. Sie können von Power BI nicht unmittelbar verarbeitet werden. Hier kommt jedoch der integrierte Query Editor ins Spiel: Dieses Tool erlaubt es dir, unstrukturierte Daten entweder manuell oder mithilfe einer automatischen Typerkennung durch Power BI in ein strukturiertes, analysierbares Format zu verwandeln.

Bevor wir tiefer in die Praxis einsteigen, schauen wir uns nun genauer an, was strukturierte und unstrukturierte Daten genau sind und worin sich diese Typen voneinander unterscheiden.

## Sturkturierte Daten

Strukturierte Daten sind Daten, die ein standardisiertes Format für den effizienten Zugriff durch Software und Menschen gleichermaßen haben. Sie sind in der Regel tabellarisch mit Zeilen und Spalten, die Datenattribute klar definieren. Computer können strukturierte Daten aufgrund ihres quantitativen Charakters effektiv verarbeiten, um Erkenntnisse zu gewinnen. Zum Beispiel kann eine strukturierte Kundendatentabelle mit den Spalten „Name“, „Adresse“ und „Telefonnummer“ Aufschluss über die Gesamtzahl der Kunden und den Ort mit der höchsten Kundenanzahl geben. 

**Definierbare Attribute**
Strukturierte Daten haben dieselben Attribute für alle Datenwerte.  Zum Beispiel könnte jeder Buchungseintrag diese Attribute haben: Name der Buchung, Ereignisname, Ereignisdatum und Buchungsbetrag.

**Relationale Attribute**
Strukturierte Datentabellen haben gemeinsame Werte, die verschiedene Datensätze miteinander verknüpfen. Zum Beispiel können Sie Kundendaten mit Buchungsdten verknüpfen, indem Sie die Felder Kunden-ID und Buchungs-ID verwenden. Sie können strukturierte Daten also einfach in einer relationalen Datenbank speichern.

**Quantitative Daten**
Strukturierte Daten eignen sich gut für mathematische Analysen. Zum Beispiel können Sie die Häufigkeit von Attributen zählen und messen, und mathematische Vorgänge auf numerischen Daten ausführen.

**Speicher**
Sie können strukturierte Daten in relationalen Datenbanken speichern und sie mit einer strukturierten Abfragesprache (SQL) verwalten. Mit SQL können Sie ein Datenmodell definieren, als Schema bekannt, in dem Sie vordefinierte Regeln – z. B. Felder, Formate und Werte – für Ihre Daten festlegen. Sie können dann strukturierte Daten in einem Data Warehouse oder mit einer anderen relationalen Datenbank-Technologie speichern.

### Beispiele von strukturierte Daten

- Excel-Dateien
- SQL-Datenbanken

## Unstrukturierte Daten

Unstrukturierte Daten sind Informationen, die in keiner fest definierten, tabellarischen Form vorliegen und somit nicht direkt in ein Datenbankschema passen. Sie erfordern in der Regel zusätzliche Verarbeitungsschritte, um nutzbar gemacht zu werden. Diese Art von Daten gewinnt zunehmend an Bedeutung, da immer mehr Informationen digital in verschiedensten Formaten erfasst werden.

### Merkmale 


- **Keine festgelegte Struktur:**
Es gibt keine klar definierten Datenschemata oder Tabellenstrukturen.

**Hohe Flexibilität:**
Unstrukturierte Daten erlauben verschiedenste Arten von Inhalten (Texte, Bilder, Videos usw.) ohne Einschränkungen bezüglich Format und Inhalt.

**Schwierige Automatisierung:**
Aufgrund fehlender Struktur ist die automatische Analyse und Verarbeitung komplex und häufig mit manueller Aufbereitung verbunden.


### Beispiele

- **CSV-Dateien (.csv):** einfache Textdateien, in denen Daten oft nur durch Kommata getrennt sind.
**PDF-Dokumente (.pdf):** enthalten oft textbasierte Informationen, die nicht unmittelbar strukturiert abgelegt sind.
**Textdateien (.txt):** bestehen aus freiem Text und benötigen spezielle Verarbeitung, um Daten extrahieren zu können.

## Einstellungen von Datenquellen

Ein wichtiges Thema in Power BI ist die Anpassung des Dateipfads bereits verbundener Datenquellen. Versucht man, den Pfad der Datenquelle direkt im Datei-Explorer oder auf Betriebssystemebene zu ändern, erkennt Power BI diese Änderung nicht automatisch. Power BI verwendet weiterhin den ursprünglich gespeicherten Pfad, wodurch es zu Fehlermeldungen beim Aktualisieren der Daten kommt.

Um dieses Problem zu beheben oder solche Fehlermeldungen von vornherein zu vermeiden, muss der Pfad direkt in Power BI geändert werden. Dazu gibt es zwei Möglichkeiten:

### Erste Möglichkeit: Über die Power BI-Menüleiste

Klicke direkt in der Power BI-Menüleiste auf den Dropdown-Pfeil neben der Schaltfläche **„Daten transformieren“**. Es öffnet sich ein Pop-up-Fenster, in dem alle verbundenen Datenquellen angezeigt werden. Wähle hier die entsprechende Datenquelle aus, deren Pfad geändert werden soll. Danach klickst du im unteren Bereich des Fensters auf die Schaltfläche **„Quelle ändern“**. Nun öffnet sich ein weiteres Fenster, in dem du den neuen Pfad festlegen kannst.

<img src="../bilder\prepare-data\lecture_1-connect_data_img_3.png" />



### Zweite Möglichkeit: Über den Query Editor

Alternativ kannst du die Änderung über den **Query Editor** vornehmen. Öffne den Query Editor und klicke in der Menüleiste auf die Schaltfläche **„Datenquelleneinstellungen“**. Es öffnet sich anschließend dasselbe Pop-up-Fenster wie bei der ersten Möglichkeit. Auch hier kannst du die gewünschte Datenquelle auswählen und den neuen Pfad einstellen.
