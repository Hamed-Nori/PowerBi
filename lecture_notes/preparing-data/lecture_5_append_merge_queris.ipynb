
# Zusammenfügen von Daten in Power BI

Ein zentraler Teil der Datenaufbereitung in Power BI besteht darin, Informationen aus mehreren Dateien oder Tabellen miteinander zu kombinieren. Häufig liegt es in unserem Interesse, Datensätze aus unterschiedlichen Quellen in einer einzigen, konsolidierten Tabelle oder „Master-Datei“ zusammenzuführen. Power BI bietet hierfür zwei Hauptmethoden, die direkt im **Query Editor** (Power Query Editor) verfügbar sind: **Append Query** und **Merge Query**.

---

## Append Query

### Was ist ein Append Query?
Ein **Append Query** (Anfügen) wird verwendet, um Datensätze zweier oder mehrerer Tabellen mit identischen Strukturen zu einer einzigen Tabelle zusammenzuführen. Mit anderen Worten: Du hast beispielsweise zwei Tabellen, die beide die gleichen Spalten enthalten (z. B. Bestellungen aus verschiedenen Monaten), und möchtest diese Tabellen zu einer Gesamt-Tabelle kombinieren.

### Wann wird ein Append Query eingesetzt?
- **Regelmäßige Daten-Ergänzung**: Wenn du regelmäßig weitere Daten erhältst, die denselben Aufbau haben, wie z. B. monatliche Verkaufsdaten.
- **Daten aus unterschiedlichen Quellen mit gleicher Struktur**: Beispielsweise mehrere CSV-Dateien, die jeweils dieselben Spalten und Felder beinhalten, aber unterschiedliche Datensätze (z. B. Kundendaten verschiedener Länder).
- **Schnelle Erweiterung einer bestehenden Tabelle**: Sobald die Struktur identisch ist, kann eine neue Datei mühelos an die bereits existierende Tabelle angehängt werden.

### Wo wird ein Append Query durchgeführt?
Die Funktion **Append Query** findest du im **Power Query Editor** von Power BI. Nachdem du deine Datenquellen in Power BI importiert hast, klicke auf den Reiter **„Daten transformieren“**, um den Query Editor zu öffnen. Dort findest du den Bereich **„Kombinieren“** (meist im oberen Menüband oder in einem Dropdown auf der rechten Seite).

### Wie wird ein Append Query ausgeführt?
1. **Datenquellen laden**: Stelle sicher, dass sowohl die ursprüngliche Tabelle als auch die hinzuzufügende Tabelle im Power Query Editor geladen sind.  
2. **Option „Abfragen anfügen“ auswählen**: Klicke in der Menüleiste auf **„Kombinieren“** → **„Abfragen anfügen“** (Append Query).  
3. **Zieltabellen auswählen**: In dem sich öffnenden Dialogfeld wählst du zunächst die „Haupttabelle“ und anschließend die Tabelle(n), die angehängt werden sollen.  
4. **Prüfen und Bestätigen**: Kontrolliere, ob die Struktur (Anzahl der Spalten und Spaltennamen) passt. Bestätige die Auswahl mit **OK**.  
5. **Ergebnis überprüfen**: Nach dem Bestätigen werden die Datensätze der angehängten Tabelle unterhalb der bestehenden Zeilen der ursprünglichen Tabelle eingefügt.

### Warum ist ein Append Query sinnvoll?
- **Zeitersparnis**: Anstatt manuell Daten zu kopieren oder mehrere Tabellen getrennt zu verwalten, kannst du sie mit wenigen Klicks zusammenführen.
- **Einheitliche Struktur**: Anhängen mehrerer Tabellen mit gleicher Struktur sorgt für eine saubere und einheitliche Datengrundlage.
- **Automatisierbarkeit**: Dank Power Query kannst du diesen Vorgang immer wieder ausführen, sodass neue Daten automatisch mit den alten Daten verbunden werden.

## Merge Query

### Was ist ein Merge Query?
Ein **Merge Query** (Zusammenführen) wird eingesetzt, wenn du Informationen aus zwei Tabellen mithilfe gemeinsamer Schlüsselspalten verknüpfen möchtest, sodass Daten „nebeneinander“ kombiniert werden. Ein typisches Beispiel wäre das Verbinden einer **Kunden-Tabelle** mit einer **Bestellungen-Tabelle**, bei der beide Tabellen über die **Kundennummer** oder **Kunden-ID** zusammengeführt werden können.

### Wann wird ein Merge Query eingesetzt?
- **Relationale Daten**: Wenn du Tabellen hast, die über einen eindeutigen Identifikator (z. B. ID) verknüpft werden können.
- **Erweitern von Datensätzen**: Falls du zusätzliche Spalten aus einer anderen Tabelle für deine Analyse benötigst (z. B. du hast Bestellinformationen, möchtest aber auch Kundendaten wie Namen oder Adressen hinzufügen).
- **Anreicherung von Daten**: Um Daten aus einer Stammdaten-Tabelle (z. B. Produktkatalog) in eine Haupttabelle (z. B. Verkaufsdaten) zu übernehmen, basierend auf Schlüsselwerten.

### Wo wird ein Merge Query durchgeführt?
Ähnlich wie beim Append Query führst du das Merge Query direkt im **Power Query Editor** durch. Im oberen Menüband oder in einem Dropdown-Menü findest du die Option **„Abfragen zusammenführen“** (Merge Query).

### Wie wird ein Merge Query ausgeführt?
1. **Datenquellen importieren**: Lade in den Power Query Editor die Tabellen, die du zusammenführen möchtest.  
2. **Option „Abfragen zusammenführen“ wählen**: Klicke auf **„Kombinieren“** → **„Abfragen zusammenführen“** (Merge Query).  
3. **Schlüsselspalten auswählen**: Markiere in beiden Tabellen die Spalten, anhand derer die Daten zusammengeführt werden sollen (z. B. Kundennummer oder Produkt-ID). Achte darauf, dass in beiden Tabellen derselbe Datentyp vorliegt und die Werte möglichst eindeutig sind.  
4. **Join-Art festlegen**: Power BI unterstützt verschiedene **Join-Arten**, die bestimmen, wie Datensätze aus den beiden Tabellen kombiniert werden.  
5. **Spalten erweitern**: Nach erfolgreichem Merge kannst du auswählen, welche Spalten der zweiten Tabelle in deiner Haupttabelle angezeigt werden sollen.  
6. **Ergebnis prüfen**: Kontrolliere die hinzugefügten Spalten und Werte, um sicherzustellen, dass das Zusammenführen korrekt funktioniert hat.

#### Die verschiedenen Join-Arten

Beim Zusammenführen der Abfragen kannst du in Power BI (ähnlich wie in SQL) den gewünschten Join-Typ auswählen:

1. **Left Outer Join**:  
   - Enthält alle Zeilen aus der ersten (linken) Tabelle sowie die passenden Zeilen aus der zweiten (rechten) Tabelle.  
   - Wenn in der rechten Tabelle keine passenden Einträge vorhanden sind, bleiben die Werte in den hinzugefügten Spalten leer (null).

2. **Right Outer Join**:  
   - Enthält alle Zeilen aus der zweiten (rechten) Tabelle sowie die passenden Zeilen aus der ersten (linken) Tabelle.  
   - Wenn in der linken Tabelle keine passenden Einträge vorhanden sind, bleiben die Werte in den hinzugefügten Spalten leer (null).

3. **Full Outer Join**:  
   - Enthält alle Zeilen aus beiden Tabellen, unabhängig davon, ob ein passender Datensatz in der jeweils anderen Tabelle existiert.  
   - Bei fehlender Übereinstimmung in der anderen Tabelle werden die entsprechenden Werte ebenfalls mit leer (null) gefüllt.

4. **Inner Join**:  
   - Enthält nur die Zeilen, die in beiden Tabellen einen passenden Eintrag haben.  
   - Datensätze, für die in der jeweils anderen Tabelle kein passender Schlüssel vorhanden ist, werden ausgeschlossen.

5. **Left Anti Join**:  
   - Enthält nur die Zeilen aus der ersten (linken) Tabelle, die in der zweiten (rechten) Tabelle **keine** Entsprechung haben.  
   - Dient dazu, Abweichungen oder fehlende Datensätze zu identifizieren.

6. **Right Anti Join**:  
   - Enthält nur die Zeilen aus der zweiten (rechten) Tabelle, die in der ersten (linken) Tabelle **keine** Entsprechung haben.

Die Wahl der richtigen Join-Art hängt von deiner konkreten Aufgabenstellung ab. Möchtest du beispielsweise nur Datensätze sehen, die in beiden Tabellen übereinstimmen, nutzt du den **Inner Join**. Willst du hingegen alle Daten aus einer Tabelle und nur die passenden Datensätze aus der anderen, verwendest du den **Left Outer Join** oder **Right Outer Join**.

### Warum ist ein Merge Query sinnvoll?
- **Zentrale Zusammenführung**: Statt in separaten Tabellen zu arbeiten, kannst du benötigte Felder aus unterschiedlichen Quellen zusammenfassen.
- **Effiziente Datenmodellierung**: Du vermeidest redundante Daten und baust ein sauberes Datenmodell auf.
- **Flexibles Daten-Matching**: Verschiedene Join-Typen ermöglichen ein gezieltes Zusammenführen, abhängig von den Anforderungen (z. B. alle übereinstimmenden Daten, nur Datensätze ohne passende IDs etc.).

## Fazit

Das **Zusammenfügen von Daten** ist ein wesentlicher Schritt in der Datenaufbereitung mit Power BI. Ob du Daten **anhängst** (**Append Query**) oder **nebeneinander zusammenführst** (**Merge Query**), der Query Editor bietet dir umfassende Möglichkeiten, Daten aus verschiedenen Quellen konsistent in einem gemeinsamen Dataset zu vereinen.

- **Append Query**: Verwende diese Methode, wenn du mehrere Dateien oder Tabellen mit **gleicher Struktur** zu einer einzigen Tabelle zusammenfassen möchtest.  
- **Merge Query**: Nutze diese Option, wenn du Daten aus **unterschiedlichen Tabellen** anhand einer **gemeinsamen Schlüsselspalte** verbinden willst, um zusätzliche Informationen spaltenweise zu integrieren. Verschiedene **Join-Arten** (Left/Right/Full Outer, Inner, Anti Join) ermöglichen dir eine sehr genaue Steuerung darüber, wie die Tabellen zusammengeführt werden.

Richtig eingesetzt schaffen beide Verfahren eine solide Grundlage für ein strukturiertes und aussagekräftiges Reporting in Power BI. Gerade bei umfangreichen oder komplexen Projekten tragen **Append** und **Merge** wesentlich dazu bei, Daten systematisch zu verwalten und effizient zu analysieren.
