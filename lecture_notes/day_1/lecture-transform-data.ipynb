# 1. Einführung in den Power Query Editor

Der Power Query Editor ist das zentrale Werkzeug in Power BI, mit dem du Daten importieren, transformieren und bereinigen kannst, bevor sie in deinen Berichten verwendet werden. Beim ersten Öffnen des Editors wirst du feststellen, dass die Benutzeroberfläche intuitiv gestaltet ist und viele Funktionen bereitstellt, um den Datenaufbereitungsprozess zu vereinfachen. Links im Fenster findest du die Navigator-Palette, in der alle geladenen Datenquellen und Abfragen übersichtlich dargestellt werden. In der Mitte befindet sich das Vorschaufenster, in dem die transformierten Daten angezeigt werden, und rechts findest du das Aufgabenfenster, das dir spezifische Optionen und Funktionen zu den aktuell ausgewählten Schritten anbietet.

Zu Beginn des Kurses ist es wichtig, die grundlegenden Komponenten des Editors zu erklären: wie man Datenquellen auswählt, wie der Importprozess funktioniert und wie die einzelnen Schritte der Transformation aufgezeichnet werden. Es ist hilfreich, den Unterschied zwischen dem Vorschaufenster und dem endgültigen Datenmodell zu betonen. Zusätzlich wird ein Einblick in den sogenannten „M“-Code gegeben, der alle Transformationen im Hintergrund beschreibt. Zwar ist ein tiefes Eintauchen in die M-Sprache für Anfänger nicht zwingend erforderlich, jedoch ermöglicht ein grundlegendes Verständnis den Teilnehmern, ihre Transformationen besser zu debuggen und zu erweitern.

Abschließend sollte demonstriert werden, wie Änderungen gespeichert und angewendet werden, sodass das finale Datenmodell in Power BI aktualisiert wird. Durch praxisnahe Beispiele und Live-Demonstrationen im Power Query Editor wird der Einstieg in die Datenaufbereitung erleichtert und die Basis für weiterführende Themen gelegt.

Ebenso spielt das Sortieren der Daten – sowohl auf- als auch absteigend – eine Rolle, um Muster und Ausreißer zu erkennen und einen besseren Überblick zu erhalten.

Auch der Umgang mit Duplikaten gehört zu den grundlegenden Transformationen. Duplizierte Einträge können die Analyse verfälschen und müssen daher erkannt und entfernt werden. Die Power Query Oberfläche bietet hierfür spezielle Funktionen. Zudem ist es wichtig, systematische Namenskonventionen zu etablieren, sodass Dateien, Spalten und Reihen stets verständlich benannt sind.

Zum Abschluss dieses Kapitels wird der Einsatz des Power Query „Schritte“-Fensters erläutert, in dem jeder Transformationsschritt dokumentiert wird. Dies erleichtert spätere Anpassungen und sorgt dafür, dass der gesamte Prozess nachvollziehbar bleibt.

# Appending Queries – Tabellen zu einer Master-Tabelle zusammenführen

Das Anhängen (Appending) von Queries ist eine leistungsstarke Methode in Power BI, um mehrere Tabellen, die im Wesentlichen den gleichen Datensatz abbilden, vertikal zu einer einzigen Master-Tabelle zusammenzuführen. Dieser Vorgang ist besonders nützlich, wenn Daten in periodischen, regionalen oder sonstigen Unterteilungen vorliegen und für eine einheitliche Analyse gebündelt werden sollen. Dabei gibt es zahlreiche wichtige Aspekte, die beachtet werden müssen, um ein konsistentes und fehlerfreies Ergebnis zu erzielen.

## Wichtige Punkte bei der Verwendung von Appending Queries

### Einheitliche Spaltennamen und Struktur
- **Konsistenz der Spaltennamen:** Alle zu kombinierenden Tabellen müssen dieselben Spaltennamen verwenden. Wenn beispielsweise in einer Tabelle die Spalte "Datum" und in einer anderen "Datum_Wert" heißt, wird der Append-Vorgang zu fehlerhaften oder nicht zusammengeführten Datensätzen führen. Es ist daher unerlässlich, die Spalten vor dem Anhängen zu überprüfen und gegebenenfalls umzubenennen.
- **Gleiche Spaltenreihenfolge und -anzahl:** Auch wenn Power BI in der Regel nur die Spalten mit übereinstimmenden Namen zusammenführt, kann es sinnvoll sein, die Tabellen vorab so zu strukturieren, dass sie exakt dieselbe Anzahl und Reihenfolge von Spalten haben. So wird sichergestellt, dass alle Daten in die korrekten Felder übernommen werden.

### Datentyp-Konsistenz
- **Homogene Datentypen:** Es ist entscheidend, dass für identische Spalten in den verschiedenen Tabellen auch dieselben Datentypen verwendet werden. Beispielsweise sollten numerische Werte in allen Tabellen als Zahl oder Dezimalzahl formatiert sein und nicht in manchen als Text. Unterschiedliche Datentypen können zu Konvertierungsfehlern oder unerwarteten Ergebnissen führen.
- **Prüfung und Anpassung vor dem Append:** Vor dem Anhängen der Abfragen sollte man sicherstellen, dass alle Spalten, die zusammengeführt werden sollen, den gleichen Datentyp besitzen. Oft ist es hilfreich, die Datenquelle oder den Power Query Editor zu nutzen, um eventuelle Unstimmigkeiten zu erkennen und zu korrigieren.

### Datenbereinigung und -vorbereitung
- **Fehlende oder zusätzliche Spalten:** Es kann vorkommen, dass nicht alle Tabellen exakt dieselbe Spaltenstruktur besitzen. Fehlende Spalten müssen entweder in der jeweiligen Tabelle ergänzt oder so behandelt werden, dass der Append-Prozess reibungslos abläuft (z.B. durch das Einfügen von leeren Werten). Ebenso sollten überflüssige Spalten, die in manchen Tabellen vorhanden sind, entfernt oder ignoriert werden.
- **Duplikate und Inkonsistenzen:** Nach dem Zusammenführen ist es ratsam, die resultierende Master-Tabelle auf Duplikate oder Inkonsistenzen zu überprüfen. Eine einheitliche Datenstruktur erleichtert nicht nur die spätere Analyse, sondern auch die Pflege und Aktualisierung der Daten.

### Automatisierung und Aktualisierungen
- **Dynamische Datenintegration:** In vielen Szenarien ist es sinnvoll, den Append-Vorgang so zu automatisieren, dass bei neuen Datenquellen oder aktualisierten Datensätzen diese nahtlos in die bestehende Master-Tabelle integriert werden. Dies kann durch Parameter oder automatisierte Aktualisierungsprozesse in Power BI erreicht werden.
- **Dokumentation und Nachvollziehbarkeit:** Jeder Schritt im Append-Prozess sollte klar dokumentiert werden, um spätere Fehler zu vermeiden und den Prozess für andere Teammitglieder nachvollziehbar zu machen. Eine saubere Dokumentation ist gerade bei großen Datenmengen und regelmäßigen Aktualisierungen unerlässlich.

Zusammenfassend lässt sich sagen, dass Appending Queries eine sehr effiziente Methode darstellt, um mehrere ähnliche Datensätze zu einer umfassenden Master-Tabelle zu kombinieren. Durch die sorgfältige Prüfung von Spaltennamen, Datentypen und einer konsistenten Struktur können Fehler vermieden werden, wodurch die Basis für fundierte Analysen in Power BI geschaffen wird.

# Merging Queries – Tabellen kombinieren und verknüpfen

Das Zusammenführen (Merging) von Queries ist ein Prozess, bei dem Daten aus mehreren Tabellen anhand gemeinsamer Schlüsselspalten horizontal kombiniert werden. Anders als beim Anhängen, bei dem Tabellen untereinander zusammengeführt werden, werden beim Merging die Inhalte nebeneinandergelegt, sodass eine erweiterte Master-Tabelle entsteht, die Informationen aus verschiedenen Quellen in einem Datensatz bündelt. Diese Methode ist ideal, um beispielsweise ergänzende Daten, wie Kunden- und Bestellinformationen, zu einer einzigen, umfassenden Ansicht zu verbinden.

## Wichtige Aspekte und zu beachtende Punkte beim Merging

### Gemeinsame Schlüsselspalten und Übereinstimmung
- **Identifikation von Schlüsselspalten:** Die Grundlage eines erfolgreichen Mergings ist die Bestimmung der gemeinsamen Schlüsselspalten. Diese Spalten enthalten eindeutige Identifikatoren (z. B. Kundennummer, Produkt-ID), die in allen zu kombinierenden Tabellen vorhanden sein müssen. Nur so kann Power BI die Datensätze korrekt zueinander in Beziehung setzen.
- **Gleiche Spaltennamen und Datentypen:** Es ist unerlässlich, dass die Schlüsselspalten in allen Tabellen denselben Namen und Datentyp haben. Unterschiedliche Bezeichnungen oder inkonsistente Formate führen dazu, dass der Merge-Vorgang nicht korrekt durchgeführt werden kann. Vor dem Zusammenführen sollten daher alle betroffenen Spalten auf Übereinstimmung geprüft und gegebenenfalls angepasst werden.

### Auswahl des richtigen Join-Typs
- **Join-Methoden und deren Auswirkungen:** Beim Merging bietet Power BI verschiedene Join-Typen an, wie den inner join, left outer join, right outer join und full outer join. Jeder dieser Join-Typen hat spezifische Auswirkungen auf das Ergebnis:
  - **Inner Join:** Es werden nur Datensätze übernommen, die in beiden Tabellen übereinstimmende Schlüsselwerte haben.
  - **Left Outer Join:** Alle Datensätze der linken Tabelle werden übernommen, und nur die passenden Einträge aus der rechten Tabelle ergänzt.
  - **Right Outer Join:** Analog zum left outer join, jedoch mit der rechten Tabelle als Basis.
  - **Full Outer Join:** Alle Datensätze aus beiden Tabellen werden kombiniert, wobei fehlende Werte als Null dargestellt werden.
- **Wahl basierend auf Analysezielen:** Die Auswahl des geeigneten Join-Typs hängt von den spezifischen Anforderungen der Analyse ab. Eine gründliche Überlegung und Testläufe sind oft notwendig, um den optimalen Merge-Vorgang zu definieren.

### Datenkonsistenz und -bereinigung
- **Vorbereitung der Datenquellen:** Wie beim Appending sollten auch beim Merging alle zu kombinierenden Tabellen vorab bereinigt werden. Dies schließt die Vereinheitlichung von Spaltennamen, die Überprüfung der Datentypen und die Beseitigung von unnötigen oder redundanten Spalten ein.
- **Behandlung von fehlenden oder doppelten Daten:** Während des Merge-Vorgangs können fehlende oder doppelte Einträge entstehen. Es ist wichtig, diese nach dem Zusammenführen zu identifizieren und zu bereinigen, um die Integrität der Master-Tabelle zu gewährleisten.
- **Validierung der Ergebnisse:** Nach dem Merge sollten die resultierenden Daten auf Konsistenz überprüft werden. Eine stichprobenartige Kontrolle der zusammengeführten Datensätze kann helfen, Fehler frühzeitig zu erkennen und zu korrigieren.

### Erweiterbarkeit und Automatisierung
- **Skalierbarkeit des Prozesses:** Merging Queries ermöglicht es, komplexe Datenbeziehungen abzubilden und mehrere Informationsquellen in einem einheitlichen Datensatz zu integrieren. Wenn neue Datenquellen hinzukommen, kann der Merge-Vorgang durch Automatisierung und Parameter dynamisch erweitert werden.
- **Dokumentation und Wiederverwendbarkeit:** Jeder Merge-Schritt sollte klar dokumentiert werden, um den Prozess nachvollziehbar zu machen. Eine gut strukturierte Dokumentation hilft nicht nur bei der Fehlerbehebung, sondern ermöglicht auch anderen Teammitgliedern, den Merge-Vorgang zu verstehen und bei Bedarf anzupassen.

Zusammenfassend stellt das Merging von Queries eine essenzielle Technik dar, um ergänzende Informationen aus verschiedenen Tabellen zu einer umfassenden Master-Tabelle zu verknüpfen. Durch die Beachtung der korrekten Schlüsselspalten, einheitlicher Spaltennamen und Datentypen sowie die sorgfältige Auswahl des Join-Typs wird eine solide Datenbasis geschaffen, die als Fundament für tiefgehende Analysen und strategische Entscheidungen in Power BI dient.
