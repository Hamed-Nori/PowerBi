# Was ist Power BI?**

Power BI ist ein von Microsoft entwickeltes Tool zur Visualisierung von Daten. Diese Software wird von zahlreichen Unternehmen verwendet, um komplexe Daten in ansprechenden und verständlichen Darstellungsformen zu präsentieren. Das Ziel ist es, aus den Daten wertvolle Erkenntnisse zu gewinnen und darauf basierende Entscheidungen zu treffen.

Um dies zu erreichen, durchlaufen die Daten mehrere Verarbeitungsschritte:

- **Transformation:**  
  Zunächst werden die Daten in eine geeignete Form gebracht, sodass sie weiterverarbeitet werden können. Häufig sind die Daten nicht in einer geeigneten Form, die für die Visualisierung möglich ist. Zum Beispiel kann es sein, dass Spaltennamen fehlen oder andere kleine Details unvollständig sind. Dafür gibt es Werkzeuge in Power BI, die uns die Möglichkeit geben, die Daten anzupassen und einen Clean-Up durchzuführen.

- **Modellierung:**  
  Anschließend werden die Daten modelliert, um Zusammenhänge herzustellen und Analysen zu ermöglichen. Nachdem wir unsere Daten transformiert haben, können wir diese modellieren. Das bedeutet, dass wir mehrere Datenquellen miteinander kombinieren können. Zum Beispiel kann es sein, dass wir mehrere Tabellen in Power BI importiert haben, die in einer Beziehung zueinander stehen oder stehen sollten.

- **Visualisierung:**  
  Zum Schluss werden die Daten in Power BI visualisiert, sodass sie in Berichten und Dashboards übersichtlich dargestellt werden. Abschließend kann anhand der Visualisierungen eine Analyse betrieben werden, um mögliche Muster zu erkennen und geeignete Maßnahmen zu ergreifen. Außerdem bietet Power BI die Möglichkeit, die visualisierten Daten zu teilen.

Unser Kurs ist so strukturiert, dass wir diese Prozesse in verschiedenen Projekten wiederholt durchlaufen. So erlernen die Teilnehmer eine einheitliche und strukturierte Arbeitsweise, die sie optimal auf die PL-300-Zertifizierungsprüfung vorbereitet. Wir beginnen mit der Transformation der Daten, gefolgt von der Modellierung, in der Daten miteinander verknüpft werden. Danach erfolgt die Auswahl der geeigneten Visualisierungsmöglichkeit, die Analyse der Daten sowie abschließend die Erstellung von Berichten, das Teilen der Ergebnisse und die Zusammenarbeit mit anderen.

Im Fokus steht Power BI Desktop. Allerdings um Power BI zu teilen und mit anderen zu arbeiten muss man sich auch mit Power BI Cloud befassen.
Es gibt drei unterteilungen von Power BI Anwendung: 

1. Power BI Desktop 
2. Power BI Cloud
3. Power BI Mobile

# Installation von power BI

**Schritte zur Installation von Power BI Desktop**

1. **Systemanforderungen überprüfen:**  
   - Stellen Sie sicher, dass Ihr Betriebssystem (Windows 10 oder höher) sowie Ihre Hardware den Mindestanforderungen entsprechen.

2. **Download von der offiziellen Microsoft-Website:**  
   - Besuchen Sie die [offizielle Power BI Desktop Download-Seite](https://powerbi.microsoft.com/desktop) und klicken Sie auf „Download“.

3. **Passende Version auswählen:**  
   - Wählen Sie die 32-Bit- oder 64-Bit-Version entsprechend Ihrer Systemarchitektur.

4. **Installationsdatei herunterladen:**  
   - Speichern Sie die heruntergeladene Datei an einem gewünschten Ort auf Ihrem Computer.

5. **Installation starten:**  
   - Doppelklicken Sie auf die heruntergeladene Datei, um den Installationsassistenten zu starten.
   - Bestätigen Sie gegebenenfalls die Benutzerkontensteuerung (UAC).

6. **Installationsanweisungen befolgen:**  
   - Akzeptieren Sie die Lizenzvereinbarung.
   - Wählen Sie den gewünschten Installationsort.
   - Klicken Sie auf „Installieren“, um den Vorgang zu beginnen.

7. **Installation abschließen:**  
   - Warten Sie, bis die Installation abgeschlossen ist.
   - Klicken Sie auf „Fertigstellen“, um den Assistenten zu beenden.

8. **Power BI Desktop starten:**  
   - Öffnen Sie Power BI Desktop über das Startmenü oder eine erstellte Desktop-Verknüpfung.

9. **Anmeldung mit Microsoft-Konto:**  
   - Melden Sie sich mit Ihrem Microsoft-Konto an, um Zugriff auf alle Funktionen zu erhalten.
