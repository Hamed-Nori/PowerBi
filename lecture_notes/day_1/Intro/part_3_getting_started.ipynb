

## 1. Der Startbildschirm – Dein erster Eindruck

Wenn du Power BI Desktop öffnest, siehst du zunächst den **Willkommensbildschirm**. Dieser Bildschirm bietet dir verschiedene Möglichkeiten:
- **Daten abrufen:** Hier kannst du direkt Datenquellen auswählen und importieren.
- **Zuletzt verwendete Quellen:** Du findest hier eine Übersicht der Datenquellen, die du zuletzt genutzt hast.
- **Berichte öffnen:** Falls du bereits mit Power BI gearbeitet hast, kannst du hier deine zuletzt erstellten Berichte öffnen.
- **Weitere Links:** Es gibt auch Links zu Tutorials und Hilfeartikeln.

Bevor du loslegst, schließe einfach den Willkommensbildschirm, indem du auf das entsprechende Symbol klickst, um in den Hauptarbeitsbereich zu gelangen.

## 2. Die Oberfläche im Überblick

### 2.1 Die Hauptansichten

Power BI Desktop gliedert sich in drei zentrale Ansichten, die dir jeweils unterschiedliche Werkzeuge zur Verfügung stellen:

- **Berichtsansicht:**  
  Hier erstellst du deine Visualisierungen und arrangierst die Berichtsseiten. Auf dem zentralen Canvas ziehst du Felder, fügst Diagramme ein und passt diese an.
  
  Ein Screenshot der Berichtsansicht zeigt dir, wie alles angeordnet ist:  
  <img src="media/desktop-getting-started/designer-gsg-report-view.png" alt="Berichtsansicht in Power BI Desktop">

- **Datenansicht:**  
  In dieser Ansicht siehst du deine geladenen Daten in Tabellenform. Du kannst hier die Daten inspizieren, Beziehungen prüfen und einfache Transformationen vornehmen.
  
- **Modellansicht:**  
  Diese Ansicht stellt dir die Beziehungen zwischen den einzelnen Tabellen grafisch dar. So kannst du sicherstellen, dass deine Datenquellen richtig miteinander verknüpft sind.

Du kannst zwischen diesen Ansichten über die Symbole auf der linken Seite wechseln. Der aktive Bereich wird durch eine farbige Markierung hervorgehoben.

### 2.2 Der Menüband-Bereich

Oben in Power BI Desktop findest du das Menüband. Hier sind alle wichtigen Befehle zusammengefasst:
- **Home:** Hier findest du Funktionen wie „Get Data“ und „Daten transformieren“.
- **Ansicht:** In diesem Bereich kannst du einstellen, welche Bereiche (wie Felder, Visualisierungen und Filter) ein- oder ausgeblendet werden sollen.
- **Einfügen, Modellieren, und weitere:** Je nach deiner Aufgabe findest du hier spezielle Werkzeuge für Visualisierungen, Datenmodellierung und Formatierung.

---

## 3. Daten importieren – Mit GET DATA starten

Um deine erste Datenquelle zu laden, klicke auf der Registerkarte **Home** auf den Button **Daten abrufen**. Es öffnet sich ein Fenster, in dem du aus vielen Datenquellen auswählen kannst.

Du kannst beispielsweise auf **Web** klicken, wenn du Daten aus dem Internet importieren möchtest. Gib die URL der gewünschten Webseite ein und bestätige mit **OK**. Anschließend zeigt dir das Navigator-Fenster die verfügbaren Tabellen an, aus denen du auswählen kannst. Wähle die Tabelle, die dich interessiert, und klicke auf **Laden** oder auf **Daten transformieren**, um die Daten vor dem Laden anzupassen.

Hier siehst du, wie der Bereich „Daten abrufen“ aussieht:  
<img src="media/desktop-getting-started/getdataweb.png" alt="Daten abrufen in Power BI Desktop">

## 4. Im Query Editor – Deine Daten bearbeiten und strukturieren

Sobald du Daten importiert hast, öffnet sich der **Power Query-Editor**. Hier kannst du deine Daten weiter bearbeiten:
- **Daten transformieren:** Du kannst Spalten umbenennen, Datentypen ändern oder unerwünschte Zeilen und Spalten entfernen.
- **Angewendete Schritte:** Alle deine Bearbeitungsschritte werden im rechten Bereich dokumentiert. So kannst du jederzeit zurückgehen und Änderungen anpassen.
- **Sortieren und Filtern:** Beispielsweise kannst du eine Spalte nach bestimmten Kriterien sortieren oder nur die Top-N-Daten anzeigen lassen.

Ein Beispiel, wie du den Datentyp änderst, siehst du hier:  
<img src="media/desktop-getting-started/designer_gsg_changedatatype.png" alt="Option zum Ändern des Datentyps in Power BI Desktop">

Nachdem du deine Anpassungen vorgenommen hast, klicke auf **Close & Apply** (Schließen und übernehmen), um die bearbeiteten Daten in den Hauptbereich von Power BI Desktop zu laden.

---

## 5. Berichte erstellen – Visualisierungen gestalten

Jetzt, wo deine Daten im System sind, geht es darum, einen ansprechenden Bericht zu erstellen.

### 5.1 Den Berichtscanvas nutzen

In der **Berichtsansicht** findest du einen großen leeren Bereich, den sogenannten Canvas. Hier ziehst du Felder aus dem Bereich **Felder** auf den Canvas, um Visualisierungen zu erstellen.  
Wenn du beispielsweise das Feld „State“ (Bundesstaat) in den Canvas ziehst, wird automatisch eine kartenbasierte Visualisierung erzeugt.

### 5.2 Visualisierungen anpassen

Im Bereich **Visualisierungen** kannst du das Aussehen deiner Diagramme verändern. Wähle das gewünschte Diagramm aus, füge weitere Datenfelder hinzu und passe das Layout sowie die Farben an.  
Du kannst auch Filter direkt im Visualisierungsbereich einstellen, um deine Daten noch aussagekräftiger darzustellen.

Hier ein Beispiel, wie du eine Visualisierung erstellen und anpassen kannst:  
<img src="media/desktop-getting-started/designer-gsg_reportfirstdrag.png" alt="Erste Visualisierung in Power BI Desktop">

### 5.3 Seiten und Layouts

Du kannst deinem Bericht mehrere Seiten hinzufügen, um verschiedene Aspekte deiner Daten getrennt darzustellen. Klicke dazu einfach auf das **+**-Symbol in der Seitenleiste oder nutze den Befehl unter **Einfügen** → **Neue Seite**.  
Seiten lassen sich leicht umbenennen und neu anordnen, sodass du einen übersichtlichen und strukturierten Bericht erstellen kannst.

Ein Beispiel der Seitenleiste findest du hier:

## 6. Deine Arbeit freigeben – Den fertigen Bericht veröffentlichen

Nachdem du deinen Bericht erstellt hast, kannst du ihn mit anderen teilen. Auch wenn wir heute die Installation und Veröffentlichung nicht im Detail behandeln, hier ein kurzer Überblick:
- Du kannst den Bericht als PBIX-Datei speichern und direkt verteilen.
- Alternativ kannst du den Bericht über den Button **Veröffentlichen** im Menüband in den Power BI-Dienst hochladen, sodass er online abrufbar ist.

## 7. Zusammenfassung und Tipps

Zusammengefasst hast du in diesem Schnellstart einen Überblick über die wichtigsten Elemente von Power BI Desktop erhalten:
- **Willkommensbildschirm:** Hier beginnst du deine Arbeit.
- **Hauptansichten:** Bericht, Daten und Modell – jede Ansicht unterstützt dich bei verschiedenen Aufgaben.
- **Get Data:** Mit diesem Werkzeug importierst du deine Daten aus diversen Quellen.
- **Query Editor:** Hier passt du die Daten an, um ein sauberes und strukturiertes Datenmodell zu erstellen.
- **Berichtsansicht:** Auf dem Canvas erstellst du deine Visualisierungen und fügst interaktive Elemente hinzu.
- **Freigabe:** Dein fertiger Bericht kann gespeichert und später veröffentlicht werden.

Ich empfehle dir, die einzelnen Funktionen auszuprobieren und mit den Werkzeugen zu experimentieren. Wenn du einmal den Umgang mit Power BI Desktop beherrscht, wirst du sehen, wie viel Spaß es macht, Daten zu visualisieren und spannende Erkenntnisse daraus zu gewinnen.

Viel Erfolg und vor allem viel Spaß beim Arbeiten mit Power BI Desktop!
